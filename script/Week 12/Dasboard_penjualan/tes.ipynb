```python
import matplotlib.pyplot as plt
import pandas as pd

# Data representasi jawaban kuesioner
data = {
    "Pernyataan": [
        "1",
        "2",
        "3",
        "4",
        "5",
        "6",
        "7"
    ],
    "SS": [14, 15, 5, 6, 10, 13, 13],
    "S":  [4, 3, 12, 10, 8, 4, 5],
    "RR": [0, 0, 0, 0, 0, 0, 0],
    "TS": [0, 0, 1, 1, 0, 1, 0],
    "STS": [0, 0, 0, 0, 0, 0, 0],
}

# Buat DataFrame
df = pd.DataFrame(data)

# Plot chart bar stacked
plt.figure(figsize=(12, 6))
bottom = [0] * len(df)
colors = {
    "SS": "#2ecc71",
    "S": "#3498db",
    "RR": "#f1c40f",
    "TS": "#e67e22",
    "STS": "#e74c3c"
}

for col in ["SS", "S", "RR", "TS", "STS"]:
    plt.bar(df["Pernyataan"], df[col], bottom=bottom, label=col, color=colors[col])
    bottom = [i + j for i, j in zip(bottom, df[col])]

plt.title("Hasil Kuesioner Pelatihan")
plt.ylabel("Jumlah Responden")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Skala")
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
