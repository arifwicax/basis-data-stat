```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Buat plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title('Contoh Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
# Data
df = pd.DataFrame({
    'kategori': ['A', 'B', 'C', 'D'],
    'nilai': [23, 45, 56, 78]
})

# Buat plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='kategori', y='nilai')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
# Data
df = pd.DataFrame({
    'tanggal': ['2024-01', '2024-02', '2024-03'],
    'penjualan': [1000, 1500, 1200]
})

# Buat plot
fig = px.line(df, x='tanggal', y='penjualan', 
              title='Tren Penjualan')
fig  # Tampilkan chart di Jupyter Notebook
```
